```python
import json


class Ticketon():
    def __init__(self):
        self.city_dict = self.read_json('city_dict')
        self.base_dict = self.read_json('base_dict')
        self.section_dict = self.read_json('section')
        
    def read_json(self,file_name):
        with open(file_name + '.json','r') as json_file:
            data = json.load(json_file)
            
        return data
    
    def get_by_city(self, city_name):
        city_events = {}
        for key, value in self.city_dict.items():
            if city_name in key:
                for event, info in value.items():
                    city_events[event] = info

        return city_events
    
    def get_by_date(self, date):
        date_events = {}
        for section, events in self.base_dict.items():
            d_events = {}
            for event, info in events.items():
                if info['time'] != None and date in info['time']:
                    date_events[event] = info

        return date_events

    def get_by_city_date(self, city_name, date):
        city_date_events = {}
        city = self.get_by_city(city_name)
        
        for city_event, info in city.items():
            if (city_event['time']!=None and date in city_event['time']):
                city_date_events[city_event] = info
                
        return city_date_events
    
    def get_by_section(self, section_name):
        section_events = {}
        for section_key, events in self.section_dict.items():
            if section_name in section_key:
                for event, info in events.items():
                    section_events[event] = info

        return section_events
    
    def get_by_city_section(self, city_name, section_name):
        city_section_events = {}
        for key, value in self.base_dict.items():
            if section_name in value:
                for event, info in value.items(): 
                    city_section_events[event] = info
                    
        return city_section_events
    
    def get_by_section_date(self, date, section_name):
        section_date_events = {}
        section = self.get_by_section(section_name)
        
        for section_event, info in section.items():
            if (section_event['time']!=None and date in section_event['time']):
                section_date_events[section_event] = info
                
        return section_date_events
    
    def get_by_city_section_date(self, city_name, section_name, date):
        city_section_date_events = {}
        city_section_dict = self.get_by_city_section(city_name, section_name)
        
        for section_event, info in city_section_dict.item():
            if (section_event['time']!=None and date in section_event['time']):
                city_section_date_events[section_event] = info
```
